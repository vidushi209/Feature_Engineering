### Nominal/OHE Encoding

one hot encoding is also known as nominal encoding, is a method used to represent categorical data as numerical data, more suitable for machine learning algorithms. In this method, each category is represented as a binary vector where each bit corresponds to a unique category.

ex: if we have categorical variable "color" with three possible values ('red','blue', 'green') we can represent it in OHE as follows:

1. red:[1,0,0]
2. green:[0,1,0]
3. blue:[0,0,1]

--> DISADVANTAGES
1. increase the complexity of data if there are numerous categories (ex: if there are 100 categories it will create 100 different binary vectors).
2. Sparse Matrix will be generated and it can lead to Overfitting

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## creating dataframe

df=pd.DataFrame({'color':['red','blue','green', 'green','red', 'blue']})
```

```python
df.head()
```

```python
## creating instance of onehotencoding

encoder=OneHotEncoder()
```

```python
## performing fit and transform

encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
encoder_df=pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df, encoder_df], axis=1)
```
