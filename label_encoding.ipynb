## Label Encoding

Label Encodingand ordinal encoding are two methods used to encode categorical data as numerical data.
Label Encoding assigns a unique numerical label to each category in the variable. The labels are usually assigned in alphabetical order or basd on the frequency of the cataegories. 

ex: if there is a "color" varible with 3 values('red', 'blue', 'green') we can represent it as:

1. red:1
2. blue:2
3. green:3

```python
import pandas as pd
df=pd.DataFrame({'color':['red','blue','green', 'green','red', 'blue']})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder=LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

## Ordinal Encoding 

Ordinal encoding is used to encode categorical data that have a intrinsic order or ranking. In this method each category is assigned a numerical value based on its position in the order.

ex: if we have categorical variable "eduaction level" with value (high svhool, college, graduate, post graduate) we can represent it as follows:

1. high school: 1
2. college: 2
3. graduate: 3
4. post graduate: 4

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({'size':['small','medium','large','medium','small','large']})
```

```python
df.head()
```

```python
ord_encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
ord_encoder.fit_transform(df[['size']])
```
