## 5 Number summary and Boxplot

```python
## minimum, maximum, Q1, Q3, IQR 
```

```python
import numpy as np
```

```python
marks=[45,32,56,75,89,90,87,67,54,45,98,99,67,74]
minimum, Q1, median, Q3, maximum=np.quantile(marks,[0, 0.25, 0.5, 0.75, 1])
```

```python
minimum, Q1, median, Q3, maximum
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-1.5*(IQR)
higher_fence=Q3+1.5*(IQR)
```

```python
lower_fence
```

```python
higher_fence
```

```python
import seaborn as sns
```

```python
sns.boxplot(marks, color='orange')
```
