## Target Guided Ordinal Encoding

This method is used to encode categorical variables based on their relationship with target variable, itis useful when we have large number of unique categories, and we want to use the data for machine learning.

In this method we replace each category in categorical variable with numerical value based on the mean or median of the target variable of the category, this creates a monotonic relationship between catgorical variable and target variable, which improves prdictive power of the model.

```python
import pandas as pd
```

```python
df=pd.DataFrame({'city':['newyork', 'london', 'paris', 'tokyo', 'newyork', 'paris'],
                'price':[200,150,300,250,180,320]})
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```

```python
df[['price', 'city_encoded']]
```
